```python
import pandas as pd
import shap
import joblib
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# Ma'lumot
df = pd.read_csv("C:/Projects/Customer_churn_project/data/processed/featured_customer_churn.csv")
X = df.drop("Exited", axis=1)

# Model
model = joblib.load("C:/Projects/Customer_churn_project/models/best_model.pkl")
model
```

```python
# Tree asosidagi modellar uchun TreeExplainer
explainer = shap.Explainer(model, X)
shap_values = explainer(X)
```

```python
# SHAP Summary Plot
shap.summary_plot(shap_values, X)
```

```python
# SHAP Feature Importance (Bar korinishida)
shap.plots.bar(shap_values)
```
