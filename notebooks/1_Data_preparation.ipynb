```python
import pandas as pd
import numpy as np
import os

df = pd.read_csv("C:\Projects\Customer_churn_project\data\customer_churn.csv")
print("Shape:", df.shape)
print("Missing values:\n", df.isnull().sum())
print("Duplicated rows:", df.duplicated().sum())
df.head()
```

```python
columns_to_drop = ["RowNumber", "CustomerId", "Surname"]
df = df.drop(columns=columns_to_drop, errors='ignore')

cat_cols = df.select_dtypes(include='object').columns
print("Categorical columns:", cat_cols)

# One-hot encoding
df = pd.get_dummies(df, drop_first=True)
```

```python
# Clean up null values
df = df.dropna()

df = df.drop_duplicates()

# Save the cleaned file
os.makedirs("C:\Projects\Customer_churn_project\data\processed", exist_ok=True)
df.to_csv("C:\Projects\Customer_churn_project\data\processed\cleaned_customer_churn.csv", index=False)
print("Cleaned dataset saved: data/processed/cleaned_customer_churn.csv")
print("Final shape:", df.shape)
```
